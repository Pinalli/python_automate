# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
# 1 - entrar no sistema da empresa
# 2 - entrar na pasta exportar
# 3 - fazer o download
# 4 - calcular indicadores(faturamento e quantidade de produtos vendidos)
# 5 - enviar e-mail para diretoria

# https://pyautogui.readthedocs.io/en/latest/quickstart.html
#!pip install pyautogui
import pyautogui  as pg #( controla mouse, teclado e tela )
import pyperclip as pc #(ignora caracteres especiais)
import pandas as pd
import time as t
# pyautogui.click -> clique
# pyautogui.write -> escrever
# pyautogui.press -> pressionar
# pyautogui.hotkey -> atalho
```

```python
pg.PAUSE = 3
# entrar no sistema da empresa
pg.hotkey('ctrl' , 't')
 # entrar na pasta exportar
pc.copy('https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing')
pg.hotkey('ctrl' , 'v')
pg.press('enter')

# t.sleep(5)
# pg.position()
# clicar no local
pg.click(x=666, y=539, clicks=2)

#fazer o download
pg.click(x=793, y=646)
pg.click(x=1595, y=232)
pg.click(x=1307, y=867)
t.sleep(2)
```

### Ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos


```python
#pip install openpyxl

import pandas as pd
table = pd.read_excel(r"C:\Users\betop\Downloads\vendas.xlsx")
display(table)

faturamento = table['Valor Final'].sum()
quantidade = table['Quantidade'].sum()

print(faturamento)
print(quantidade)
```

### Vamos agora enviar um e-mail pelo gmail

```python
pg.hotkey('ctrl' , 't')
 # entrar na pasta exportar
pc.copy('https://mail.google.com/mail/u/0/?ogbl#inbox')
pg.hotkey('ctrl' , 'v')
pg.press('enter')
t.sleep(4)

pg.click(x=137, y=220)

pg.write("betopinalli@hotmail.com")
pg.press('tab')
pg.press('tab')

pc.copy("Relatório de Vendas")
pg.hotkey('ctrl' , 'v')
pg.press('tab')

pc.copy(f"""Prezados, bom dia!
O faturamento de ontem foi de R${faturamento: .2f}
A quantidade de produtos vendidos foi de {quantidade: .}
""")
pg.hotkey('ctrl' , 'v')

pg.hotkey('ctrl' , 'enter')
```
